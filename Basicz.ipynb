Import libraries to run fft

```python
%matplotlib inline
import seaborn
import numpy, scipy, matplotlib.pyplot as plt, librosa, IPython.display, mir_eval, urllib
```

Pull audio file from MIR website

```python
#filename = 'simple_loop.wav'
filename = '125_bounce.wav'
#filename = 'conga_groove.wav'
#filename = '58bpm.wav'
url = 'http://audio.musicinformationretrieval.com/' + filename
```

Retrieve url file, and import into folder

```python
urllib.urlretrieve(url, filename=filename)
```

```python
%ls *.wav
```

```python
# x recieves array of numpy floating points
# fs recieves 'sr' value -- 'None' uses native sampling rate of file
x, fs = librosa.load(filename, sr=None)
```
